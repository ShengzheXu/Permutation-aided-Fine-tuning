```python
state_shapefile_path = "../data/us_states_shapefile/States_shapefile.shp"
us_states = gpd.read_file(state_shapefile_path)
bd = {}
for state_code in us_states['State_Code'].unique():
    state_geometry = us_states[us_states['State_Code'] == state_code]['geometry'].iloc[0]
    bd[state_code] = state_geometry
```

```python
def calc_violation_rate(df, shape, df_state_name='state_code'):
    vio = 0
    for index, row in df.iterrows():
        if not shape[row[df_state_name]].contains(gpd.points_from_xy([row['lon']], [row['lat']]))[0]:
            vio += 1
    violation_rate = vio / len(df)

    return violation_rate
```

```python
from sdmetrics.reports.single_table import QualityReport

report = QualityReport()
```

```python
def gen_report(df_synth, df_real):
    # make a primary key column with index
    df_synth['primary_key'] = df_synth.index
    df_real['primary_key'] = df_real.index
    # make column order the same
    df_synth = df_synth[df_real.columns]

    metadata = {"primary_key": "primary_key", 
                "columns": {
                    "primary_key": { 
                        "sdtype": "id",
                        "regex_format": "[0-9]"
                    }
                }
                }
    for column in df_real.columns:
        if column == 'primary_key':
            continue
        if df_real[column].dtype == 'float64':
            metadata["columns"][column] = {"sdtype": "numerical", "compute_representation": "Float"}
        elif df_real[column].dtype == 'int64':
            metadata["columns"][column] = {"sdtype": "numerical", "compute_representation": "'Int64'"}
        else:
            metadata["columns"][column] = {"sdtype": "categorical"}

    report.generate(df_real, df_synth, metadata)

    return report
```

```python
# for loop files in ./paft
column_shape_record = [['data','ctgan', 'copulagan', 'tabsyn', 'great', 'paft']]
column_pair_trend_record = [['data','ctgan', 'copulagan', 'tabsyn', 'great', 'paft']]

for file in os.listdir('./data_to_eval/paft'):
    print(f'Processing {file}')
    # test_file remove after last _
    t_file = file.split("_")[:-1]
    t_file = "_".join(t_file)

    df_real = pd.read_csv(f'./data_to_eval/test/{t_file}.csv')
    df_paft = pd.read_csv(f'./data_to_eval/paft/{file}')
    df_great = pd.read_csv(f'./data_to_eval/great/{file}')
    df_ctgan = pd.read_csv(f'./data_to_eval/ctgan/{file}')
    df_copulagan = pd.read_csv(f'./data_to_eval/copulagan/{file}')
    df_tabsyn = pd.read_csv(f'./data_to_eval/tabsyn/{file}')

    report_paft = gen_report(df_paft, df_real)
    column_paft, pair_paft = report_paft.get_properties()['Score'].to_list()

    report_great = gen_report(df_great, df_real)
    column_great, pair_great = report_great.get_properties()['Score'].to_list()

    report_ctgan = gen_report(df_ctgan, df_real)
    column_ctgan, pair_ctgan = report_ctgan.get_properties()['Score'].to_list()

    report_copulagan = gen_report(df_copulagan, df_real)
    column_copulagan, pair_copulagan = report_copulagan.get_properties()['Score'].to_list()

    report_tabsyn = gen_report(df_tabsyn, df_real)
    column_tabsyn, pair_tabsyn = report_tabsyn.get_properties()['Score'].to_list()


    column_shape_record.append([file, column_ctgan, column_copulagan, column_tabsyn, column_great, column_paft])
    column_pair_trend_record.append([file, pair_ctgan, pair_copulagan, pair_tabsyn, pair_great, pair_paft])  

   
    
df_column_shape = pd.DataFrame(column_shape_record[1:], columns=column_shape_record[0])
df_column_pair_trend = pd.DataFrame(column_pair_trend_record[1:], columns=column_pair_trend_record[0])
```

```python
def check_one(d):
    vio = 0
    for index, row in d.iterrows():
        if not bd[row['state_code']].contains(gpd.points_from_xy([row['lon']], [row['lat']]))[0]:
            vio += 1
    rate = vio / len(d)

    return rate

def vio_by_state(df):
    states = df['state_code'].unique()
    vio_dict = []
    for state in states:
        vio_rate = check_one(df[df['state_code'] == state])
        vio_dict.append([state, vio_rate])
    return vio_dict
```

```python
# for loop files in ./paft
paft_by_state_vio = [['state', 'vio']]
great_by_state_vio = [['state', 'vio']]
ctgan_by_state_vio = [['state', 'vio']]
copulagan_by_state_vio = [['state', 'vio']]
tabsyn_by_state_vio = [['state', 'vio']]

for file in os.listdir('./data_to_eval/paft'):
    if not file.startswith('geo_bird_latlevel'):
        continue
    print(f'Processing {file}')
    # test_file remove after last _
    t_file = file.split("_")[:-1]
    t_file = "_".join(t_file)

    df_real = pd.read_csv(f'./data_to_eval/test/{t_file}.csv')
    df_paft = pd.read_csv(f'./data_to_eval/paft/{file}')
    df_great = pd.read_csv(f'./data_to_eval/great/{file}')
    df_ctgan = pd.read_csv(f'./data_to_eval/ctgan/{file}')
    df_copulagan = pd.read_csv(f'./data_to_eval/copulagan/{file}')
    df_blob_tabsyn = pd.read_csv(f'./data_to_eval/tabsyn/{file}')

    vio_dict_paft = vio_by_state(df_paft)
    vio_dict_great = vio_by_state(df_great)
    vio_dict_ctgan = vio_by_state(df_ctgan)
    vio_dict_copulagan = vio_by_state(df_copulagan)
    vio_dict_tabsyn = vio_by_state(df_blob_tabsyn)

    paft_by_state_vio.extend(vio_dict_paft)
    great_by_state_vio.extend(vio_dict_great)
    ctgan_by_state_vio.extend(vio_dict_ctgan)
    copulagan_by_state_vio.extend(vio_dict_copulagan)
    tabsyn_by_state_vio.extend(vio_dict_tabsyn)

```
